```python
faces = ['H', 'T']
```

```python
import random
```

```python
random.choice(faces)
```

```python
coins = random.choices(faces, k=20)
print(coins)
```

```python
coins.count('H')
```

```python
num_sims = 20
num_flips = 20
for sim in range(num_sims):  # The simulation loop
    # Simulate all coin flips for one experiment
    coins = random.choices(faces, k=num_flips)
    print(coins.count('H'), end=' ')
```

```python
num_sims = 100
num_flips = 20
results = []
for sim in range(num_sims):  # The simulation loop
    # Simulate all coin flips for one experiment
    coins = random.choices(faces, k=num_flips)
    results += [coins.count('H')]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(range(num_sims), results);
```

```python
plt.scatter(range(num_sims), results)
plt.xlabel('Experiment number')
plt.ylabel('Number of heads observed')
plt.title('Simulation of no. of heads observed on 20 coin flips');
```

```python
plt.hist(results)
```

```python
plt.hist(results, bins=15);
```

```python
import numpy as np

edges = np.arange(1.5, 15.5, 1)
print(edges)
```

```python
plt.hist(results, bins=edges);
```

```python
plt.hist(results, bins=edges, edgecolor='black')
plt.yticks(np.arange(0, 22, 2))
plt.grid(axis='y')
plt.xlabel('Number of Heads')
plt.ylabel('Number of occurrences')
```

```python
plt.hist(results, bins=edges, edgecolor='black', density=True)
```
